3. **Research question**

### What is the Strength and Mass Transport of the Northward Somali Current during September 2024?

The western boundary current plays a crucial role in the Arabian Sea circulation, serving as an indicator of the gyre’s strength, and is recognized as a significant upwelling system worldwide due to the volume of wind-induced upwelling water. The Somali Current reverses direction in response to the monsoons, which is unique among large boundary currents. The upwelling influences the moisture that drives monsoon rainfall. Model results suggest that Somali Current upwelling will intensify with climate warming, potentially leading to lower surface water temperatures. However, logistical challenges in conducting oceanographic fieldwork mean that much remains to be explored in this remote region. Our ambition, therefore, is to measure and observe the Somali Current on a research vessel during September, which is the month with the highest expected velocities, and to observe the Great Whirl gyre on our northernmost transect. For this, we use a CTD (measuring salinity and temperature over depth) and a high-resolution ADCP (measuring the flow velocities). This research aims to measure and calculate the mass transport and velocities of the Somali Current and the gyre.
